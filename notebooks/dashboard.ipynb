```python
import numpy as np
import pandas as pd
```

```python
file_path = '../raw_data/resale-flat-prices-based-on-registration-date-from-jan-2017-onwards.csv'
df_ = pd.read_csv(file_path, parse_dates = ['month'])
df = df_.copy()
```

```python
df.drop(columns=['block'],inplace=True)
```

```python
df['r_lease'] = df['remaining_lease'].apply(lambda x: (int(x[:2])) if x[9:11]== '0 ' or x[9:11] == '' else (int(x[:2]) + int(x[9:11])/12))
df.drop(columns = ['remaining_lease'], inplace = True)
```

```python
df.columns
```

```python
df.rename(columns={'month':'Month','town':'Town','flat_type':'Flat Type',
          'street_name':'Street Name','storey_range':'Storey Range',
          'floor_area_sqm':'Floor Area (Sqm)','flat_model':'Flat Model',
          'r_lease':'Remaining Lease (Years)','resale_price':'Resale Price (SGD)',
           'lease_commence_date':'Lease Commencement Year'}, inplace=True)
```

```python
df['quarter_year']=df['Month'].dt.year
df['quarter_month']=df['Month'].dt.month.apply(lambda x: 1 if x in [1, 2, 3] else 2 if x in [4, 5, 6] else 3 if x in [7, 8, 9] else 4)
df['Quarter']= df.apply(lambda row: f"{row['quarter_year']} Q{row['quarter_month']}", axis=1)
df.drop(columns=['quarter_year','quarter_month'], inplace=True)
```
